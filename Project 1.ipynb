```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

```python
df.describe() #data of columns containing numeric values
```

```python
df.info() #datatypes of columns
```

```python
df.isnull().sum() #to identify numver of null values
```

# Drop unnamed column

```python
df = df.drop("Unnamed: 0", axis =1)
print(df.head())
```

# change weekly study hours column

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","5-10") #data transformation
df.head()
```

# gender distribution

```python
plt.figure(figsize = (4,4.5))  #customization of height, length of chart
ax = sns.countplot(data = df, x ='Gender')
ax.bar_label(ax.containers[0])  #ax used to exactly show the count of genders
plt.title("Gender Distribution")
plt.show
```

```python
#from the above chart we have analyzed that:
#the number of females in the data is moore than the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb, annot = True)  #heatmap(seaborn) is used to show data in chart
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

```python
#from the above chart we have concluded that the education of the parents have good impact on students score
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb1, annot = True)  
plt.title("Relationship between Parent's Marital Status and Student's Score")
plt.show()
```

```python
#from the above chart we have concluded that the parents marital status 
#have no/nefligible impact on students score
```
